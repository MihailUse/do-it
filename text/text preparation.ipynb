```python
import re
from string import punctuation, digits, whitespace
from tqdm import tqdm
```

```python
from pymystem3 import Mystem

mystem = Mystem() 
```

```python
def clean_text(text):
    text=text.lower() 
    text=re.sub(r'[0-9a-z]|(\W+|\s+)|[\_]', ' ',text)
    text=re.sub(r'[\s]{2,}', ' ',text)
    return text
```

```python
def mystem_preprocessor(doc):
    res = []    
    for text in tqdm(doc):
        # очистка слов
        text = clean_text(text)
        
        # перевод слов к словарной корневой форме
        tokens = mystem.lemmatize(text)
        
        text = []
        for token in tokens:
            if token not in (punctuation, digits, whitespace) and len(token) > 2:
                text.append(token.strip())
        text = " ".join(text)
        res.append(text)
    return res
```

```python
%%time 

text = mystem_preprocessor(doc)
print(len(text), len(text[0]))
text
```

Более быстрый способ

```python
def mystem_preprocessor(doc):
    res = []    
    # очистка слов        
    res = [clean_text(text) for text in doc]
        
    # объединение текстов в один большой текст
    res = ' | '.join(res)+' | '
    tokens = mystem.lemmatize(res)
    
    # разделение текста и очистка от повторяющихся слов
    res = []
    text = []
    for token in tqdm(tokens):
        if token.strip() == '|':
            res.append(' '.join(text))
            text = []
        elif token not in (punctuation, digits, whitespace) and len(token) > 2:
            text.append(token.strip())

    return res
```

```python
%%time 

text = mystem_preprocessor(doc)
print(len(text), len(text[0]))
text
```
