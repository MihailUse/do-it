В этом блоке мы раззмотрим модели для машинного обучения

```python
import pandas as pd
from string import punctuation
import re
```

```python
df = pd.read_csv('train.csv', index_col=['PassengerId'])
```

```python
df
```

```python
# Дополняем данные
exdata = [
    {
        'Survived' : 1,
        'Pclass' : 2,
        'Name' : 'Unknown',
        'Sex' : 'male',
        'Age' : 20.0,
        'SibSp' : 0,
        'Parch' : 0,
        'Ticket' : 'NaN',
        'Fare' : 7.2500,
        'Cabin' : 'NaN',
        'Embarked' : 'S'
    }
]
```

```python
# Функция добавления данных
df = df.append(exdata, ignore_index=True)
df
```

```python
# Функция очистки от пунктуации(from string import punctuation)
def clean_text(text):
    for symbol in punctuation:
        text = text.replace(symbol, '')
    return text
```

```python
# Приминяем функцию ко всему датасету, но здесь нам это не надо, поэтому мы используем следующий блок 
# for col in df:
#     df[col] = df[col].map(clean_text)
```

```python
# Приминяем очистку текста от пунктуации к одному столбцу
df['Name'] = df['Name'].map(clean_text)
```

```python
df
```

```python
# Функция очистки
def clean_df(inpdf):
    inpdf.drop(['Name'], axis=1, inplace=True)
    inpdf.drop(['SibSp'], axis=1, inplace=True)
    inpdf.drop(['Ticket'], axis=1, inplace=True)
    inpdf.drop(['Embarked'], axis=1, inplace=True)
    inpdf.drop(['Parch'], axis=1, inplace=True)
    inpdf.drop(['Cabin'], axis=1, inplace=True)

# Заполняем все неизвестные значения на ноль
    df.fillna(0, inplace=True)
    
# Создаём словарь    
    sexiq = dict(zip(inpdf['Sex'].unique(), range(inpdf['Sex'].nunique())))

# Приминяем словарь
    inpdf['Sex'].replace(sexiq, inplace=True)
        
    return inpdf
```

```python
df = clean_df(df)
```

```python
df
```

```python
# Рзаделения датасета на тестовую и тренирующую моедль
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.drop('Survived', axis=1), df['Survived'], test_size=0.3)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from catboost import CatBoostClassifier
from sklearn import svm
from sklearn.linear_model import RidgeClassifierCV
from tqdm import tqdm
import pickle
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Самая лучшая модель. Может обучиться на чём угодно
rfr = RandomForestClassifier(n_estimators=1000, max_depth=20)
%time rfr.fit(x_train, y_train)
print('Случайный лес: ',rfr.score(x_test, y_test), '\n')
```

```python
# Ищет соседей
neigh = KNeighborsClassifier(n_neighbors=1)
%time neigh.fit(x_train, y_train)
print('Neighbor: ', neigh.score(x_test, y_test))
```

```python
# Очень хорошая модель, но лишь до того момента, как не встретит большые данные
cat = CatBoostClassifier(n_estimators=1)
%time cat.fit(x_train, y_train)
print('Категоризированный бустинг: ',cat.score(x_test, y_test), '\n')
```

```python
# Та же фишка с катекарайзером
gbr = GradientBoostingClassifier(n_estimators=100, max_depth=2)
%time gbr.fit(x_train, y_train)
print('Градиентный бустинг: ',gbr.score(x_test, y_test), '\n')
```

```python
# Средняя модель, но если под рукой нет других, то очень удобно
ada = AdaBoostClassifier(n_estimators=100)
%time ada.fit(x_train, y_train)
print('Adaboosting: ', ada.score(x_test, y_test))
```

```python
# Крутой наивный бейсоковский. Ищет вероятность выбранного класса
navvay = GaussianNB()
%time navvay.fit(x_train, y_train)
print('Beans: ', navvay.score(x_test, y_test))
```

```python
# Наивный бейсоковский. Ищет вероятность выбранного класса
mnb = MultinomialNB()
%time mnb.fit(x_train, y_train)
print('MNB: ', mnb.score(x_test, y_test))
```

```python
# Линейный дискриминантный анализ. Лучше использовать, когда данные имеют линейную зависимость. Работает за счёт уменьшения размерностинабра данных, проецируя все точки данных на линию.
lda = LinearDiscriminantAnalysis()
%time lda.fit(x_train, y_train)
print('Точность линейного дискременанта: ', lda.score(x_test, y_test))
```

```python
# Машины поддержки векторов работают, рисуя линию между различными кластерами точек данных, чтобы сгруппировать их в классы
clf = svm.SVC()
%time clf.fit(x_train, y_train)
print('Машины опорных векторов: ', clf.score(x_test, y_test))
```

```python
# Просто линейная модель
rcv = RidgeClassifierCV()
%time rcv.fit(x_train, y_train)
print('Линейная модель: ', rcv.score(x_test, y_test))
```

```python
# Перекодировка слов в числа
encod = {}
proc_df = df.copy()
str_df = df.select_dtypes(object)

for col in str_df:
    encod[col] = LabelEncoder()
    proc_df[col] = encod[col].fit_transform(str_df[col])
proc_df
```
